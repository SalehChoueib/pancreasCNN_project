## Project and Aims
**This project aims to develop and validate a CNN-based to predict non-invasively predict malignancy from CT scans.** 


**Impact**:
   
       This research will provide a validated molecular-clinico-pathomic-radiomic risk score to optimally select patients at high risk of pancreatic cancer who would benefit from resection, while preventing the over-treatment of patients with benign disease. Enhanced non-invasive diagnostic tools will improve early detection and prevention of pancreatic cancer, an incurable disease.


**Importance of the Problem & Critical Barrier**: 

       The failure of current diagnostic and therapeutic strategies for pancreatic cancer is exemplified by the fact that even for patients with resectable disease (<15% of all pancreatic cancer patients), the probability of long-term survival is quite low. Of 985 patients who underwent resection for “localized” pancreatic cancer at Memorial Sloan Kettering Cancer Center (MSK) between 1983 and 2005, the majority had stage IIB disease (nodal metastases, 61% of cases), and just 28 patients (3% of resectable cases) had the earliest stage (stage IA) pancreatic cancer. These patients, with the earliest lesions that can be identified by either cross-sectional imaging or endoscopy, experienced a dismal five-year survival of 35%. Since 1983, two patients who have undergone resection for pancreatic cancer and survived more than 20 years. These survival rates suggest that a critical barrier is identification of pancreatic cancer at a curable stage and preventing overtreatment of benign cystic lesions

## Data 

#### CT Volume Data: 

    The data we are working with for this project is the pre-operative abdominal CT scans of 104 patients. Each of these abdominal CT volumes were segmented twice, once in regards to the pancreas, and again in regards to the cyst in the pancreas. These segmentations were saved as separate CT volumes, though they correspond to the original abdominal CT. 

    In the current pipeline, the segmented cysts are being used both to select which slices from the abdominal CT volumes to execute a patch extraction and to center the cropping of the CT slice into a 64x64 image. 
    
#### Clinical Data:

    The CT volumes have corresponding clinical data in a .xlsx file, that contains labels, patient factors, and cyst factors.
